## SOLUTION Q2: Completa la seguente implementazione della rete bayesiana usando la libreria CausalGraphicalModels

```python
# NB Update 19/12/2025 to guarantee the compatibility

!pip install causalgraphicalmodels
#!sed -i 's/collections/collections.abc/g' /usr/local/lib/python3.10/dist-packages/causalgraphicalmodels/cgm.py
!sed -i 's/collections/collections.abc/g' /usr/local/lib/python3.12/dist-packages/causalgraphicalmodels/cgm.py
from causalgraphicalmodels import CausalGraphicalModel
```

```python
offer = CausalGraphicalModel(
    nodes=["Experience", "Grades", "Interview", "Offer"],
    edges=[
        ("Experience", "Interview"),
        ("Grades", "Interview"),
        ("Interview", "Offer"),
    ]
)

offer.draw()
```

## SOLUTION Q3: Supponendo che si potesse fare un intervention sull'Experience, come si implementerebbe usando la libreria CausalGraphicalModels

```python
new_offer = offer.do("Experience")

new_offer.draw()
```
